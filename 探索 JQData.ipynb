## 获取所有基金列表

```python
# 获取现在还在上市的基金
fund_df = get_all_securities(types='fund', date=datetime.now())
fund_df
```

```python
# 确保退市日期均为 2200（没有退市）
assert len(set(fund_df.end_date.values)) == 1

# 提取上市日期
funds = fund_df.start_date.items()
```

## 获取基金累计净值

```python
# 提取一个基金
fund = next(funds)

# 显示基金中文名称
print(get_security_info(fund[0]).display_name)

# 获取基金自上市至 2018/9/1 的累计净值
fund_value = get_extras('acc_net_value', [fund[0]], start_date=fund[1], end_date='2018-9-1')
plt.plot(fund_value)
```
